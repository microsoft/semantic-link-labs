```python
import sempy_labs as labs
import sempy_labs.lakehouse as lake
import sempy_labs.admin as admin
```

# List Items in a Workspace

Using the admin list items function, filter down the list of returned items to only be lakehouses. 
[Docs on admin.list_items](link-URL)

```python
# List all items from a workspace

workspaceName = 'Semantic Link Labs'


lakehouseList = admin.list_items(workspace=workspaceName, type='lakehouse')
lakehouseList
```

```python
# Select the lakehouse by row index
lakehouseName = lakehouseList['Item Name'][1]

# Alternatively you could save the name into the lakehouse name
# lakehouseName = 'Your_Lakehouse_Name'

lakehouseName
```

```python
# List tables from the selected lakehouse

lakehouseTables = lake.get_lakehouse_tables(lakehouse=lakehouseName)
lakehouseTables
```

# Optimize

**PySpark Optimize** refers to a set of techniques and tools used to improve the performance of PySpark applications. It includes query optimization through [Catalyst](https://www.databricks.com/glossary/catalyst-optimizer), physical execution planning via Tungsten, and efficient memory and resource management. These optimizations help reduce execution time, minimize data shuffling, and enhance overall scalability in distributed data processing


```python
# Run Optimize on a single table

## Run the optimization on the first table returned from the list of tables captured above
lake.optimize_lakehouse_tables(lakehouse=lakehouseName, tables=lakehouseTables['Table Name'][0])
```

```python
# Alternatively you can optimize a single table by name as Text

## Enter the name as text
lake.optimize_lakehouse_tables(lakehouse=lakehouseName, tables='Your_Table_Name')
```

```python
# Run Optimize on a list of tables

## using different syntax we can pass a list of tables into the Optimize function
## 0:2 selects the range of tables starting at row 0 and 1 and not 2.
## Note indexing starts at 0 for the first row of data, also you must add the .tolist() 
##    function to convert the series of tables into a list

lake.optimize_lakehouse_tables(lakehouse=lakehouseName, tables=lakehouseTables['Table Name'][0:2].tolist())
```

```python
# Run optimize on the entire lakehouse

lake.optimize_lakehouse_tables(lakehouse=lakehouseName)
```

# Vacuum Tables

**PySpark Vacuum** is a maintenance operation used primarily with Delta Lake tables to clean up obsolete data files. It removes files that are no longer referenced by the Delta transaction log, helping to free up storage and improve performance. By default, it retains data for a set period (e.g., 7 days) to ensure compatibility with time travel and data recovery features.

```python
# run Vacuum on one table

lake.vacuum_lakehouse_tables(lakehouse=lakehouseName, tables='Your_Table_Name')
```

```python
# run Vacuum on a list of tables

lake.vacuum_lakehouse_tables(lakehouse=lakehouseName, tables=lakehouseTables['Table Name'][0:2].tolist())
```

```python
# Run Vacuum on the entire lakehouse

lake.vacuum_lakehouse_tables(lakehouse=lakehouseName)
```
